```python
import pandas as pd
```

```python
amzn = pd.read_csv('amazon_prime_titles.csv')
dsny = pd.read_csv('disney_plus_titles.csv')
ntlx = pd.read_csv('netflix_titles.csv')
```

```python
pf = pd.DataFrame(['amazon','disney','netflix'],columns=['platform_name']).reset_index().rename(columns={'index':'platform_id'})
pf
```

```python
pf.to_csv('platforms.csv',index=False)
```

```python
amzn.show_id +='a'
dsny.show_id +='d'
ntlx.show_id +='n'
```

```python
amzn['platform_id'] = 0
dsny['platform_id'] = 1
ntlx['platform_id'] = 2
```

```python
amzn_cast = get_cast_df(amzn)
dsny_cast = get_cast_df(dsny)
ntlx_cast = get_cast_df(ntlx)
```

```python
big_cast_df = pd.concat([amzn_cast,dsny_cast,ntlx_cast],ignore_index=True)
big_cast_df
```

```python
big_cast_df.to_csv('casts.csv',index=False)
```

```python
amzn_director = get_cast_df(amzn,'director')
dsny_director = get_cast_df(dsny,'director')
ntlx_director = get_cast_df(ntlx,'director')

big_director_df = pd.concat([amzn_director,dsny_director,ntlx_director],ignore_index=True)
big_director_df
```
